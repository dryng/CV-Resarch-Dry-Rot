```python
import matplotlib.pyplot as plt
import numpy as np
import json
import os
import seaborn as sns
sns.set_theme()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(10,5)
folders = os.listdir("classification")
metrics = {}
for folder in folders:
    file=f"classification/{folder}/metrics.json"
    with open(file,'r') as f:
        metrics[folder] = json.load(f)
        ax.plot(list(range(len(metrics[folder]['val_losses']))),metrics[folder]['val_losses'],label=folder)

ax.set_title('Validation loss for classification models')
ax.set_xlabel('Epoch')
ax.set_ylabel('BCE')
ax.legend()
# plt.savefig("validation_loss.jpg")
plt.show()
```
